```python
# creating our own data type fraction

class Fraction:

  def __init__(self,x,y):
    self.num = x
    self.den = y

  # By default printing method
  def __str__(self):
    return "{}/{}".format(self.num,self.den)

  # Object come here by default for addition
  def __add__(self,other):
    new_num = self.num*other.den + other.num*self.den
    new_den = self.den*other.den

    return "{}/{}".format(new_num,new_den)

  # Object come here by default for subtraction
  def __sub__(self,other):
    new_num = self.num*other.den - other.num*self.den
    new_den = self.den*other.den

    return "{}/{}".format(new_num,new_den)

  # Object come here by default for multiplication
  def __mul__(self,other):
    new_num = self.num * other.num
    new_den = self.den*other.den

    return "{}/{}".format(new_num,new_den)

  # Object come here by default for division
  def __truediv__(self,other):
    new_num = self.num*other.den
    new_den = self.den*other.num

    return "{}/{}".format(new_num,new_den)

  # our own method for converting fraction into decimal
  def decimal(self):
    return self.num/self.den
```

**Let's create our own range() function**

```python
class mera_range:

    def __init__(self,start,end):
        self.start = start
        self.end = end

    def __iter__(self):
        return mera_range_iterator(self.start,self.end)
```

```python
class mera_range_iterator(mera_range):

    def __init__(self,start,end):
      super().__init__(start , end)
      self.start = start
      self.end = end

    def __iter__(self):
        return self

    def __next__(self):

        if self.start >= self.end:
            raise StopIteration

        current = self.start
        self.start+=1
        return current
```

```python
# generator function
import sys
def gen_func():
  for i in range(1,11):
    yield i

gen = gen_func()
print(sys.getsizeof(gen))
for i in gen:
  print(i)
```

```python
# Genrator square func

def square(num):
  for i in range(1,num+1):
    yield i**2


gen = square(10)
print(sys.getsizeof(gen)/64)
for i in gen:
  print(i)
```
