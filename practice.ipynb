```python
# creating our own data type fraction

class Fraction:

  def __init__(self,x,y):
    self.num = x
    self.den = y

  # By default printing method
  def __str__(self):
    return "{}/{}".format(self.num,self.den)

  # Object come here by default for addition
  def __add__(self,other):
    new_num = self.num*other.den + other.num*self.den
    new_den = self.den*other.den

    return "{}/{}".format(new_num,new_den)

  # Object come here by default for subtraction
  def __sub__(self,other):
    new_num = self.num*other.den - other.num*self.den
    new_den = self.den*other.den

    return "{}/{}".format(new_num,new_den)

  # Object come here by default for multiplication
  def __mul__(self,other):
    new_num = self.num * other.num
    new_den = self.den*other.den

    return "{}/{}".format(new_num,new_den)

  # Object come here by default for division
  def __truediv__(self,other):
    new_num = self.num*other.den
    new_den = self.den*other.num

    return "{}/{}".format(new_num,new_den)

  # our own method for converting fraction into decimal
  def decimal(self):
    return self.num/self.den
```
